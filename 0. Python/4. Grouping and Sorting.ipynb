```python
import pandas as pd

df = pd.read_csv('data/winemag-data-130k-v2.csv')
df.head()
```

```python
# Group by country and get the minimum price
df.groupby('country').price.min()

# Group by winery and get the first title
df.groupby('winery').apply(lambda x: x.title.iloc[0])
```

```python
df. (['country', 'province']).apply(lambda x: x.loc[x.points.idxmax()])
```

```python
# Group by country and get the mean, median, min, and max price
df.groupby('country').price.agg([len, 'mean', 'median', min, max])
```
